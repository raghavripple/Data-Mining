```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/daily_sales.csv')

Q1 = df['Sales Amount'].quantile(0.25)
Q3 = df['Sales Amount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Sales Amount'] < lower_bound) | (df['Sales Amount'] > upper_bound)]
print("Detected outliers:")
print(outliers)

plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Sales Amount'])
plt.title("Boxplot of Sales Amount")
plt.show()

lower_cap = df['Sales Amount'].quantile(0.05)
upper_cap = df['Sales Amount'].quantile(0.95)
df['Sales Amount'] = np.clip(df['Sales Amount'], lower_cap, upper_cap)

df.to_csv('cleaned_daily_sales.csv', index=False)
print("Cleaned dataset saved as 'cleaned_daily_sales.csv'")
```
